# Train T5 model

```
# pip install simpletransformers
```

```
# verificamos si hay GPU disponible
print(torch.cuda.is_available())
```

```
from simpletransformers.t5 import T5Model, T5Args


model_args = T5Args()
model_args.num_train_epochs = 3


cuda_available = torch.cuda.is_available()
model = T5Model(
    model_type="t5",
    model_name="t5-small",
    use_cuda = cuda_available,
    args=model_args
)
```

```
model.train_model(train_data, output_dir ='.Model/t5_1')
```

```
to_predict = [
    "paraphrase: horror movie"
]

predictions = model.predict(to_predict)
```

Load trained Model

Nota: cambiar el model_name con la dirección donde se encuentra el modelo

```
from simpletransformers.t5 import T5Model, T5Args
import torch

model_args = T5Args()
model_args.do_sample = True
model_args.top_p = 0.7
model_args.num_return_sequences = 2

cuda_available = torch.cuda.is_available()
model = T5Model(
    model_type="t5",
    model_name="/content/drive/MyDrive/IAG/Obligatorio2/Modelo 6/outputs",
    use_cuda = cuda_available,
    args=model_args
)
```

```
to_predict = [
    "paraphrase: horror movie. ",
    "paraphrase: Mica my friend.",
    "paraphrase: This is an amazing film",
"paraphrase: Action movie",
"paraphrase: Scary movie",
"paraphrase: Hello how are you",
"paraphrase: Im cold",
"paraphrase: This is cool",
"paraphrase: I like to play football",
"paraphrase: Shall we go in?",
"paraphrase: My dog is huge",
"paraphrase: Hello! How's life treating you these days?"
]

predictions = model.predict(to_predict)
```
