Esta notebook contiene el preprocesamiento de los datos

El modelo debe recibir 3 features: el prefijo(prefix), el input_text y el output text

# Imports

```python
import pandas as pd
```

### tapaco dataset

```python

tapaco_dataset = pd.read_csv("tapaco_paraphrases_dataset.csv",sep="\t")
tapaco_dataset.head()
```

```python
# Rename columns
tapaco_dataset.rename(columns={'Text': 'input_text', 'Paraphrase': 'target_text'}, inplace=True)
```

```python
# Add prefix column
tapaco_dataset['prefix'] = 'paraphrase'
```

```python
# Reorder the columns
tapaco_dataset = tapaco_dataset[['prefix', 'input_text', 'target_text']]
tapaco_dataset.head()
```

```python
# Download csv file
tapaco_dataset.to_csv("datasets/tapaco_dataset.csv", index=False)
```

### Paws dataset

```python
paws_dataset = pd.read_csv("paws_dataset.csv", sep=",")
paws_dataset.head()
```

```python
# Rename columns
paws_dataset.rename(columns={'sentence1': 'input_text', 'sentence2': 'target_text'}, inplace=True)
```

```python
# Add prefix and reorder the columns
paws_dataset['prefix'] = 'paraphrase'
paws_dataset = paws_dataset[['prefix', 'input_text', 'target_text']]
```

```python
paws_dataset.head()
```

```python
paws_dataset.to_csv("datasets/paws_dataset.csv", index=False)
```

### Combine datasets

```python
result = pd.concat([tapaco_dataset,paws_dataset], ignore_index=True)
```

```python
result.shape
```

### Extended dataset

```python
result.to_csv('datasets/datasets_combine.csv', index=False)
```
